```python
K0 = 10000 #Startkapital
T = 40 #Laufzeit
Z = 200 #Jährliche Auszahlung
mu = 0.02 #Erwartung der jährlichen Renditen
sigma = 0.15 #Varianz der jährlichen Renditen
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_paths(N):
    K = np.zeros((N, T+1))
    
    K[:, 0] = K0
    
    
    # Number of normal draws needed
    
    n = N * T
    
    
    if n % 2 == 1:
    
        n += 1
    
    
    # Uniform random numbers # U cannot be zero, but with machine error nothing goes wrong here
    
    U1 = np.random.uniform(0, 1, n//2)
    
    U2 = np.random.uniform(0, 1, n//2)
    
    
    # Box-Muller transform
    
    Z1 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * np.pi * U2)
    
    Z2 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * np.pi * U2)
    
    
    # trimming
    
    Z_all = np.concatenate([Z1, Z2])[:N*T]
    
    
    # Scale to desired normal distribution
    
    R = sigma * Z_all + mu
    
    
    # Reshape into (n paths × T years)
    
    R = R.reshape(N, T)
    
    
    for t in range(T):
    
        K[:, t+1] = K[:, t] * (1 + R[:, t]) - Z
    return K

for i in range(20):
    plt.plot(get_paths(20)[i], alpha=0.6)

plt.title("Simulated Capital Paths (Box-Muller)")
plt.xlabel("Time (Years)")
plt.ylabel("Capital")
plt.show()
```

```python
multiplicator = 1
N = 2000
Nar = multiplicator*(np.arange(N)+1)   
mean = np.zeros(N)
v = np.zeros(N)

for i in range(N):
    mean[i] = get_paths(Nar[i])[:,-1].mean()
    v[i] = np.sqrt(get_paths(Nar[i])[:,-1].var())

    
f, (sub1, sub2) = plt.subplots(2,1,sharex=True)
sub1.scatter(Nar, mean, s=0.2)
sub1.set_ylabel('Mean of K_T')
sub2.scatter(Nar, v, s=0.2)
sub2.set_xlabel('Number of paths')
sub2.set_ylabel('St. deviation of K_T');
```
