```python
K0 = 10000 #Startkapital
T = 40 #Laufzeit
Z = 300 #Jährliche Auszahlung
mu = 0.02 #Erwartung der jährlichen Renditen
sigma = 0.15 #Varianz der jährlichen Renditen
n = 10
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
K = np.zeros((n,T))
K0n = np.array([K0 for i in range(n)])
K = np.c_[K0n, K]

for k in range(n):
    for i in range(T):
        U1, U2 = np.random.uniform(0,1,2)
        x1 = np.sqrt(-2*np.log(U1))*np.cos(2*np.pi*U2)
        R = x1*sigma + mu
        K[k][i+1] = K[k][i]*(1+R) - Z
    plt.plot(K[k])
    

```
