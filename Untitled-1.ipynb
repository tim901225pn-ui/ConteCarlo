```python
K0 = 10000 #Startkapital
T = 40 #Laufzeit
Z = 300 #Jährliche Auszahlung
mu = 0.02 #Erwartung der jährlichen Renditen
sigma = 0.15 #Varianz der jährlichen Renditen
n = 20
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
K = np.zeros((n, T+1))
K[:, 0] = K0

# Number of normal draws needed
N = n * T

if N % 2 == 1:
    N += 1

# Uniform random numbers
U1 = np.random.uniform(0, 1, N//2)
U2 = np.random.uniform(0, 1, N//2)

# Box-Muller transform
Z1 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * np.pi * U2)
Z2 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * np.pi * U2)

# trimming
Z_all = np.concatenate([Z1, Z2])[:n*T]

# Scale to desired normal distribution
R = sigma * Z_all + mu

# Reshape into (n paths × T years)
R = R.reshape(n, T)

for t in range(T):
    K[:, t+1] = K[:, t] * (1 + R[:, t]) - Z

for i in range(20):
    plt.plot(K[i], alpha=0.6)

plt.title("Simulated Capital Paths (Box-Muller)")
plt.xlabel("Time (Years)")
plt.ylabel("Capital")
plt.grid(True)
plt.show()
```
